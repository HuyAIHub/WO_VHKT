```python
def day_to_month(df_):
  #drop phong ban
  df_ = df_.drop('phong_ban',axis=1)
  # Đảm bảo cột 'Date' là kiểu datetime
  df_['date'] = pd.to_datetime(df_['date'])
  # Resample theo tháng và tính tổng giá trị
  return df_.resample('M', on='date').sum()
```

#build model

```python
learn_r = 30
training_rate = 0.8

def data_input_model(data,training_rate,learn_r):
  df = pd.DataFrame(data,columns=['date','total_duration'])
  df.index = df.date
  df.drop('date',axis=1,inplace=True)

  #chia tập dữ liệu
  rate = round(training_rate*df.shape[0])
  data = df.values

  train_data = data[:rate]
  test_data = data[rate:]
  #chuẩn hóa dữ liệu
  sc = MinMaxScaler(feature_range=(0,1))
  sc_train = sc.fit_transform(data)

  #tạo vòng lặp các giá trị
  x_train,y_train=[],[]
  for i in range(learn_r,len(train_data)):
    x_train.append(sc_train[i-learn_r:i,0]) #lấy 50 giá đóng cửa liên tục
    y_train.append(sc_train[i,0]) #lấy ra giá đóng cửa ngày hôm sau

  #xếp dữ liệu thành 1 mảng 2 chiều
  x_train = np.array(x_train)
  y_train = np.array(y_train)

  #xếp lại dữ liệu thành mảng 1 chiều
  x_train = np.reshape(x_train,(x_train.shape[0],x_train.shape[1],1))
  y_train = np.reshape(y_train,(y_train.shape[0],1))
  return x_train, y_train , train_data, test_data , data, df , sc, sc_train
```

```python
def build_model(x_train,y_train,epoch,batch_size,save_model):
  #xây dựng mô hình
  model = Sequential() #tạo lớp mạng cho dữ liệu đầu vào
  #2 lớp LSTM
  model.add(LSTM(units=128,input_shape=(x_train.shape[1],1),return_sequences=True))
  model.add(LSTM(units=64))
  # model.add(LSTM(units=64))
  model.add(Dropout(0.5)) #loại bỏ 1 số đơn vị tránh học tủ (overfitting)
  model.add(Dense(1)) #output đầu ra 1 chiều
  #đo sai số tuyệt đối trung bình có sử dụng trình tối ưu hóa adam
  model.compile(loss='mean_absolute_error',optimizer='adam')
  #huấn luyện mô hình
  best_model = ModelCheckpoint(save_model,monitor='loss',verbose=2,save_best_only=True,mode='auto')
  model.fit(x_train,y_train,epochs=epoch,batch_size=100,verbose=2,callbacks=[best_model])
  return model
```

```python
#processing data input
x_train1, y_train1 , train_data1, test_data1 , data1, df1, sc1, sc_train1 = data_input_model(df_days['df_wo_co_dien_day'],training_rate,learn_r)
x_train2, y_train2 , train_data2, test_data2 , data2, df2, sc2, sc_train2 = data_input_model(df_days['df_wo_ksnn_day'],training_rate,learn_r)
x_train3, y_train3 , train_data3, test_data3 , data3, df3, sc3, sc_train3 = data_input_model(df_days['df_wo_noc_day'],training_rate,learn_r)
x_train4, y_train4 , train_data4, test_data4 , data4, df4, sc4, sc_train4 = data_input_model(df_days['df_wo_others_day'],training_rate,learn_r)
```

```python
def evaluate(x_train,y_train,train_data,data,df,sc,final_model,name_wo):
  #dữ liệu train
  y_train = sc.inverse_transform(y_train) #giá thực
  y_train_predict = final_model.predict(x_train) #dự đoán giá đóng cửa trên tập đã train
  y_train_predict = sc.inverse_transform(y_train_predict) #giá dự đoán

  #xử lý dữ liệu test
  test = df[len(train_data)-learn_r:].values
  test = test.reshape(-1,1)
  sc_test = sc.transform(test)

  x_test = []
  for i in range(learn_r,test.shape[0]):
    x_test.append(sc_test[i-learn_r:i,0])
  x_test = np.array(x_test)
  x_test = np.reshape(x_test,(x_test.shape[0],x_test.shape[1],1))
  x_test

  #dữ liệu test
  rate = round(training_rate*df.shape[0])
  y_test = data[rate:] #giá thực
  y_test_predict = final_model.predict(x_test)
  y_test_predict = sc.inverse_transform(y_test_predict) #giá dự đoán

  #lập biểu đồ so sánh
  train_data1 = df[learn_r:rate]
  test_data1 = df[rate:]

  plt.figure(figsize=(24,8))
  plt.plot(df,label='Thời gian thực tế',color='red') #đường giá thực
  train_data1['Dự đoán'] = y_train_predict #thêm dữ liệu
  plt.plot(train_data1['Dự đoán'],label='Thời gian dự đoán train',color='green') #đường giá dự báo train
  test_data1['Dự đoán'] = y_test_predict #thêm dữ liệu
  plt.plot(test_data1['Dự đoán'],label='Thời gian dự đoán test',color='blue') #đường giá dự báo test
  plt.title('So sánh Thời gian dự báo và Thời gian thực tế của ' + name_wo) #đặt tên biểu đồ
  plt.xlabel('Thời gian') #đặt tên hàm x
  plt.ylabel('total duration') #đặt tên hàm y
  plt.legend() #chú thích
  plt.show()

  #r2
  print('Độ phù hợp tập train:',r2_score(y_train,y_train_predict))
  #mae
  print('Sai số tuyệt đối trung bình trên tập train (VNĐ):',mean_absolute_error(y_train,y_train_predict))
  #mae
  print('Phần trăm sai số tuyệt đối trung bình tập train:',mean_absolute_percentage_error(y_train,y_train_predict))

  #r2
  print('Độ phù hợp tập test:',r2_score(y_test,y_test_predict))
  #mae
  print('Sai số tuyệt đối trung bình trên tập test (time):',mean_absolute_error(y_test,y_test_predict))
  #mae
  print('Phần trăm sai số tuyệt đối trung bình tập test:',mean_absolute_percentage_error(y_test,y_test_predict))
```

```python
# evaluate(x_train1, y_train1 , train_data1 , data1, df1, sc1,final_model1,'wo_co_dien')
evaluate(x_train2, y_train2 , train_data2 , data2, df2, sc2,final_model2,'wo_knss')
evaluate(x_train3, y_train3 , train_data3 , data3, df3, sc3,final_model3,'wo_noc')
evaluate(x_train4, y_train4 , train_data4 , data4, df4, sc4,final_model4,'wo_others')
```
